```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configration of the plots we will create

#Read in the data

df = pd.read_csv(r'C:\Users\kriti\Downloads\movies.csv')
```

```python
#let's look at the data

df.head()
```

```python
#Checking missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'. format(col, pct_missing))
```

```python
#Data types for our column

df.dtypes
```

```python
#Change data types of columns

df['budget'] = df['budget'].astype("Int64")

df['gross'] = df['gross'].astype("Int64")
```

```python
df.head()
```

```python
#Create correct release column

df['YearCorrect'] = df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
#

df = df.sort_values(by=['gross'], inplace=False, ascending=False )
```

```python
pd.set_option('display.max_rows',None)
```

```python
#drop duplicates if any

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
#Budget high correlation
#company high correalation

#scatter plot with budget vs gross

df.plot(kind='scatter', x='budget', y='gross')

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for films')

plt.show()
```

```python
df.head()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# let's look at correlation

df.corr(numeric_only=True, method='pearson')
```

```python
#correlation matrix

correlation_matrix = df.corr(numeric_only=True, method='pearson')
plt.title('Correlation matric for numeric features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
